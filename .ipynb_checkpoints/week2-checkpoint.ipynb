# Week 2: Data Visualization & EDA
## Objective
To visualize the distribution of sensitive attributes (gender, race), analyze income distribution across groups, and identify potential dataset imbalance using visual tools

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv("adult.csv")
df.head()
```

## Gender Distribution

```python
gender_counts = df["sex"].value_counts()

plt.figure(figsize=(6,4))
ax = sns.countplot(x="sex", data=df)

for i in ax.containers:
    ax.bar_label(i)

plt.title("Gender Distribution")
plt.show()


```

## Race Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(x="race", data=df)
plt.xticks(rotation=45)
plt.title("Race Distribution")
plt.show()
```

```python
## Overall Income Distribution
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="income", data=df)
plt.title("Overall Income Distribution")
plt.show()
```

## Income Distribution by Gender

```python
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="income", data=df)
plt.title("Income Distribution by Gender")
plt.show()
```

```python
income_gender = pd.crosstab(df["sex"], df["income"], normalize="index") * 100
income_gender
```

## Income Distribution by Race

```python
plt.figure(figsize=(8,5))
sns.countplot(x="race", hue="income", data=df)
plt.xticks(rotation=45)
plt.title("Income Distribution by Race")
plt.show()
```

## Correlation Heatmap of Numerical Features

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df["age"].hist(bins=30)
plt.title("Age Distribution")
plt.show()
```

## Observations

1. The dataset shows gender imbalance with significantly more males than females.
2. A higher proportion of males earn >50K compared to females.
3. The race distribution is dominated by one group, indicating potential dataset imbalance.
4. Income distribution is skewed toward <=50K, suggesting class imbalance.
5. Numerical features show moderate correlations, with education and hours-per-week potentially influencing income.
